```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrames
instagram_df = pd.DataFrame({
    'Candidates': ['Daniel Noboa', 'Luisa Gonzales'],
    'Amount of Instagram Shares': [18060.0, 436.0],
    'Percentage': [97.64, 2.36]
})

tiktok_df = pd.DataFrame({
    'Candidates': ['Daniel Noboa', 'Luisa Gonzales'],
    'TikTok Saves': [28068.0, 20918.0],
    'Percentage': [57.3, 42.7]
})

facebook_df = pd.DataFrame({
    'Candidates': ['Daniel Noboa', 'Luisa Gonzales'],
    'Facebook Shares': [41310.0, 14799.0],
    'Percentage': [73.62, 26.38]
})

youtube_df = pd.DataFrame({
    'Candidate': ['Luisa Gonzales', 'Daniel Noboa'],
    'Number of Likes': [53.0, 37.0],
    'Percentage': [58.89, 41.11]
})

# Function to display DataFrames with styling
def display_styled_df(df, title):
    styled_df = df.style.set_table_attributes('style="width: 80%; margin: auto;"') \
        .set_table_styles([{
            'selector': 'thead th',
            'props': [('background-color', '#f0f0f0'), ('color', '#333'), ('font-weight', 'bold')]
        }, {
            'selector': 'tbody td',
            'props': [('background-color', '#ffffff'), ('color', '#333')]
        }, {
            'selector': 'tbody tr:nth-child(even)',
            'props': [('background-color', '#f9f9f9')]
        }])
    
    display(styled_df)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Display each DataFrame
display_styled_df(instagram_df, "Instagram Shares Data")
display_styled_df(tiktok_df, "TikTok Saves Data")
display_styled_df(facebook_df, "Facebook Shares Data")
display_styled_df(youtube_df, "YouTube Likes Data")
```

```python
import pandas as pd

# Create the DataFrames
instagram_df = pd.DataFrame({
    'Candidates': ['Daniel Noboa', 'Luisa Gonzales'],
    'Amount of Instagram Shares': [18060.0, 436.0],
    'Percentage': [97.64, 2.36]
})

tiktok_df = pd.DataFrame({
    'Candidates': ['Daniel Noboa', 'Luisa Gonzales'],
    'TikTok Saves': [28068.0, 20918.0],
    'Percentage': [57.3, 42.7]
})

facebook_df = pd.DataFrame({
    'Candidates': ['Daniel Noboa', 'Luisa Gonzales'],
    'Facebook Shares': [41310.0, 14799.0],
    'Percentage': [73.62, 26.38]
})

youtube_df = pd.DataFrame({
    'Candidate': ['Luisa Gonzales', 'Daniel Noboa'],
    'Number of Likes': [53.0, 37.0],
    'Percentage': [58.89, 41.11]
})

# Sum the percentages from each DataFrame
total_percentage = (
    instagram_df['Percentage'].sum() +
    tiktok_df['Percentage'].sum() +
    facebook_df['Percentage'].sum() +
    youtube_df['Percentage'].sum()
)

print(f"Total Percentage: {total_percentage:.2f}")
```

```python
import matplotlib.pyplot as plt

# Data
candidates = ['Daniel Noboa', 'Luisa Gonzales']
percentages = [67.42, 32.58]

# Create bar chart
plt.figure(figsize=(8, 5))  # Set the figure size
bars = plt.bar(candidates, percentages, color=['#6A0DAD', '#1E90FF'], edgecolor='black')

# Add title and labels
plt.title('Predicciones Electorales Ecuador 13 de Abril 2025', fontsize=16, fontweight='bold')
plt.xlabel('Candidatos', fontsize=14)
plt.ylabel('Porcentaje (%)', fontsize=14)

# Add percentage labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}%', ha='center', va='bottom', fontsize=12)

# Show grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Set y-axis limit
plt.ylim(0, 100)

# Show the chart
plt.tight_layout()  # Adjust layout to make room for labels
plt.show()
```
